```python
import numpy as np
```

```python
conditions = [
    df_ecommerce['TotalAmount'] < 5000,
    (df_ecommerce['TotalAmount'] >= 5000) & (df_ecommerce['TotalAmount'] < 10000),
    df_ecommerce['TotalAmount'] >= 10000
]
choices = ['Inexpensive','Moderate','Expensive']
df_ecommerce['Price_tier'] = np.select(conditions,choices,default='Unknown')
```

```python
conditions = [
    df_ecommerce['Rating']<2.5,
    (df_ecommerce['Rating']>=2.5) & (df_ecommerce['Rating']<4),
    df_ecommerce['Rating']>=4
]
choices = ['Low','Good','Excellent']
df_ecommerce['Rating_tier'] = np.select(conditions,choices,default='Unknown')
```

```python
conditions = [
    df_ecommerce['Reviews'] == 0,
    (df_ecommerce['Reviews'] >= 1000) & (df_ecommerce['Reviews'] < 2500),
    df_ecommerce['Reviews'] >= 2500
]
choices = ['Negative','Neutral','Positive']
df_ecommerce['Reviews_tier'] = np.select(conditions,choices,default='Unknown')
```
